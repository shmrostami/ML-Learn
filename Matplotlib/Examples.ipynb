```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Line Plot

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]

plt.plot(years, iran_pop)
plt.show()
```

```python
plt.figure(figsize=(11, 8), dpi=100)
plt.plot(years, iran_pop, color='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Iran\'s Population')
plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],['21M','28M','38M','56M',
'66M','74M','80M'])
plt.xticks([1960, 1970, 1980, 1990, 2000, 2010, 2016] )
plt.show()
```

#### Scatter Plot

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]

plt.scatter(years, iran_pop)
plt.show()
```

```python
popsize = np.array(iran_pop)*3
plt.scatter(years, iran_pop, s=popsize)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Iran\'s Population')
plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],['21M','28M','38M','56M',
'66M','74M','80M'])
plt.xticks([1960, 1970, 1980, 1990, 2000, 2010, 2016] )
plt.show()
```

#### Multiple Plot

```python
turkey_pop = [27.47, 34.88, 43.98, 53.92, 63.24, 72.33, 79.51]

plt.plot(years, iran_pop,ls='-',lw=2,marker='+',mew=8)
plt.plot(years, turkey_pop,ls='--',lw=1)
plt.title('Iran vs Turkey Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([20,30,40,50,60,70,80],['20M', '30M', '40M', '50M' ,'60M', '70M', '80M',])
plt.legend(['iran','Turkey'],loc='best')
#plt.text(1980,40,'IranIraq War',fontsize=15)
plt.grid()
plt.annotate('IranIraq War',xy=(1980,38.67) ,xytext=(1990,34.88) ,arrowprops=dict(facecolor='silver',width=4),fontsize=15)
plt.show()
```
