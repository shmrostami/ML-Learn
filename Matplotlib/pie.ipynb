```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

\
**Example 1**

```python
l = ['Python', 'Java' , 'C++']
s = [  200   ,  150   ,  100]
c = [ 'green', 'red', 'yellow']
```

```python
plt.pie(s,  labels=l, colors=c);
```

```python
plt.pie(s,  labels=l, colors=c, autopct='%1.1f%%');
```

```python
plt.pie(s,  labels=l, colors=c, startangle=120 );
```

```python
plt.pie(s,  labels=l, colors=c, explode=(0.1, 0, 0), shadow=True, autopct='%1.1f%%');
```

**Example 2**

```python
s = pd.Series(3 * np.random.rand(3), index=['a', 'b', 'c'], name='series')
s
```

```python
s.plot.pie();
```

```python
s.plot.pie(labels=['AA', 'BB', 'CC'], colors=['b', 'g', 'r'], fontsize=20,figsize=(6, 6));
```
